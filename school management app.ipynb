# Sample AI Generated report for the whole Teacher Dataset

```python
Generate Descriptive Statistics
mean_punctuality = df['Punctuality Score'].mean()
median_punctuality = df['Punctuality Score'].median()
mean_feedback = df['Student Feedback Score'].mean()
median_feedback = df['Student Feedback Score'].median()
mean_class_performance = df['Class Performance Score'].mean()
median_class_performance = df['Class Performance Score'].median()

mean_attendance_rate = df['Teacher Attendance Rate'].mean()

numeric_columns = df.select_dtypes(include='number')
correlations = numeric_columns.corr()

Prepare Insights
insights = f"""
The average punctuality score among teachers is {mean_punctuality:.2f}, with a median score of {median_punctuality}. 
This indicates that most teachers are quite punctual, with minimal variance in scores.

On the feedback side, the average student feedback score is {mean_feedback:.2f}, showing that students generally have 
positive feedback about their teachers. The median feedback score of {median_feedback} supports this trend.

The class performance scores average at {mean_class_performance:.2f}, suggesting that the overall student performance 
in classes is good but has room for improvement, as indicated by the lower end of the score spectrum.


There is a positive correlation between punctuality and student feedback scores, indicating that punctual teachers tend 
to receive better feedback. Similarly, a moderate correlation between professional development hours and class performance 
scores suggests that teachers who engage more in professional development tend to have better-performing classes.
"""

# Step 5: Generate a narrative using NLP
summarizer = pipeline("summarization")

# Summarize the insights
summary = summarizer(insights, max_length=150, min_length=50, do_sample=False)

# Step 6: Generate the final report
report = f"""
**Teacher Performance Report: XYZ School**

**Summary of Findings:**

{summary[0]['summary_text']}

**Detailed Insights:**

1. **Punctuality:** The average punctuality score is {mean_punctuality:.2f}, with teachers showing consistent timeliness.
2. **Student Feedback:** Teachers receive an average feedback score of {mean_feedback:.2f}, indicating positive student interactions.
3. **Class Performance:** The average class performance score is {mean_class_performance:.2f}, suggesting a need for targeted improvements.

4. **Teacher Attendance:** High attendance rates with an average of {mean_attendance_rate:.2f}%.

**Correlations and Insights:**

- Positive correlation between punctuality and feedback scores.
- Moderate correlation between professional development and class performance.

**Recommendations:**

- Encourage continued professional development to enhance class performance.
- Focus on maintaining high punctuality as it correlates with positive student feedback.
- Implement targeted strategies to improve overall class performance scores.
"""

print(report)
```

# Sample AI Generated report for the specific teacher

```python
import pandas as pd
from transformers import pipeline
import pandas as pd
from transformers import pipeline
import csv

df = pd.read_excel('teacher_perf_eval (1).xlsx')

df.to_csv('your_file.csv', index=False)


# Step 2: Generate Descriptive Statistics for the dataset
overall_stats = {
    'mean_punctuality': df['Punctuality Score'].mean(),
    'mean_feedback': df['Student Feedback Score'].mean(),
    'mean_class_performance': df['Class Performance Score'].mean(),
    'mean_prof_dev_hours': df['Professional Development Hours'].mean(),
    'mean_attendance_rate': df['Teacher Attendance Rate'].mean()
}

# Step 3: Function to generate a personalized report for each teacher
def generate_teacher_report(teacher_id):
    teacher = df[df['Teacher_ID'] == teacher_id].iloc[0]  # Get the teacher's row
    insights = f"""
Teacher Report for {teacher['Teacher_ID']} - {teacher['Subject']}

**Overview**:
- Gender: {teacher['Gender']}
- Race/Ethnicity: {teacher['Race/Ethnicity']}
- Highest Level of Education: {teacher['Highest Level of Education']}
- Years of Experience: {teacher['Years of Experience']}
- Subject: {teacher['Subject']}

**Punctuality**:
- Punctuality Score: {teacher['Punctuality Score']}
- Compared to the average punctuality score of {overall_stats['mean_punctuality']:.2f}, this teacher is {'above' if teacher['Punctuality Score'] > overall_stats['mean_punctuality'] else 'below'} average.

**Student Feedback**:
- Student Feedback Score: {teacher['Student Feedback Score']}
- Compared to the average feedback score of {overall_stats['mean_feedback']:.2f}, the feedback from students is {'positive' if teacher['Student Feedback Score'] > overall_stats['mean_feedback'] else 'average or needs improvement'}.

**Class Performance**:
- Class Performance Score: {teacher['Class Performance Score']}
- The class performance is {'above' if teacher['Class Performance Score'] > overall_stats['mean_class_performance'] else 'below'} the overall average score of {overall_stats['mean_class_performance']:.2f}.

**Professional Development**:
- Professional Development Hours: {teacher['Professional Development Hours']}
- This teacher has completed {teacher['Professional Development Hours']} hours of professional development, which is {'above' if teacher['Professional Development Hours'] > overall_stats['mean_prof_dev_hours'] else 'below'} the average of {overall_stats['mean_prof_dev_hours']:.2f} hours.

**Attendance**:
- Attendance Rate: {teacher['Teacher Attendance Rate']}%
- The teacher's attendance rate is {'higher' if teacher['Teacher Attendance Rate'] > overall_stats['mean_attendance_rate'] else 'lower'} than the average attendance rate of {overall_stats['mean_attendance_rate']:.2f}%.
"""
    
    # Step 4: Generate a narrative using NLP
    summarizer = pipeline("summarization")
    
    # Summarize the insights
    summary = summarizer(insights, max_length=150, min_length=50, do_sample=False)
    
    # Step 5: Generate the final report
    report = f"""
**Teacher Performance Report for {teacher['Teacher_ID']}**

**Summary of Findings:**

{summary[0]['summary_text']}

**Detailed Insights:**

{insights}
"""
    return report

# Step 6: Generate report for a specific teacher (e.g., Teacher with ID 'T001')
teacher_report = generate_teacher_report('T001')
print(teacher_report)
```

# Sample AI generated Report for the specific student from the collected data

```python
import pandas as pd
from transformers import pipeline

# Step 1: Load the dataset
df = pd.read_csv('StudentsPerformance (2).csv')  # Assuming the dataset is in CSV format

# Step 2: Generate overall statistics for comparison
overall_stats = {
    'mean_math': df['math score'].mean(),
    'mean_reading': df['reading score'].mean(),
    'mean_writing': df['writing score'].mean(),
}

# Step 3: Function to generate personalized report for each student
def generate_student_report(student_id):
    student = df.iloc[student_id]
    
    # Detailed Insights
    insights = f"""
Student Report for {student['gender']} student from {student['race/ethnicity']}

**Overview**:
- Gender: {student['gender']}
- Race/Ethnicity: {student['race/ethnicity']}
- Parental Education: {student['parental level of education']}
- Lunch Type: {student['lunch']}
- Test Preparation Course: {student['test preparation course']}

**Scores**:
- Math Score: {student['math score']}
    - The student's math score of {student['math score']} is {'above' if student['math score'] > overall_stats['mean_math'] else 'below'} the overall average of {overall_stats['mean_math']:.2f}.
- Reading Score: {student['reading score']}
    - The reading score of {student['reading score']} is {'above' if student['reading score'] > overall_stats['mean_reading'] else 'below'} the overall average of {overall_stats['mean_reading']:.2f}.
- Writing Score: {student['writing score']}
    - The writing score of {student['writing score']} is {'above' if student['writing score'] > overall_stats['mean_writing'] else 'below'} the overall average of {overall_stats['mean_writing']:.2f}.

**Analysis**:
- This student {('benefited from' if student['test preparation course'] != 'none' else 'did not participate in')} a test preparation course.
- Students with a {student['parental level of education']} typically perform {'better' if student['math score'] > overall_stats['mean_math'] else 'average or below'} in math.
"""
    
    # Step 4: Summarize insights using NLP
    summarizer = pipeline("summarization")
    
    summary = summarizer(insights, max_length=150, min_length=50, do_sample=False)
    
    # Step 5: Final Report
    report = f"""
**Student Performance Report**

**Summary**:
{summary[0]['summary_text']}

**Detailed Insights**:
{insights}
"""
    return report

# Step 6: Generate the report for a specific student
student_report = generate_student_report(0)  # Report for the first student
print(student_report)
```
